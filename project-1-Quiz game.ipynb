```python
q1 = """Who's the golden maknae in kpop?
a.mina
b.jk
c.joy
d.lisa"""

q2 = """what's the taylor shift latest album?
a.Midnights
b.music of Spheres
c.map of the soul
d.the Album"""

q3 = """what's the one of the best character in anime? 
a.levi
b.gojo
c.goku
d.l/light yagami"""

q4 = """what's the best anime?
a.Deathnote
b.one piece
c.attack on titan
d.Dragon ball"""

q5 = """what's the best movie?
a.the harry potter series
b.Avatar series
c.fast and furious
d.inception"""

questions = {q1:'b',q2:'a',q3:'c',q4:'b',q5:'a'}
name = input("Hello, what's your name?")
print("Hi",name,"Welcome to the Quiz!!")
score = 0 
for i in questions:
    print(" ")
    print(i)
    flag = input("Do you want to Skip the Question! (yes/no): ")
    if flag == "yes":
        continue
    ans = input("enter your answer: ")
    if ans == questions[i]:
        print("Wow!!, you got it correct")
        score = score+1 
        print("your current score is: ",score)
    else:
        print("Wrong answer, you lost 1 point")
        score = score-1
        print("your current score is: ",score)
    flag1 = input("Do u want to Quit? (yes/no) ")
    if flag1 == "yes":
        break
print("your final score: ",score)
    
```
